```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def president_list(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')

    name = []
    for i in soup.find_all('h3'):
        name.append(i.text)
    
    term = []
    for i in soup.find_all('p'):
        term.append(i.text)
    
    office_term = term[0:-3]
    
    index = [1,3,5,7]
    for i in sorted(index, reverse = True):
        del office_term[i]
    
    ind = []
    for i in range(1,15):
        ind.append(i)
    
    former_presidents = pd.DataFrame({'Name': name, 'Term of Office': office_term }, index = ind)
    return former_presidents
    
    
    
    
```

```python
url = 'https://presidentofindia.nic.in/former-presidents.htm'
a = president_list(url)
a
```
