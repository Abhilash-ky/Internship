```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
def latest(url):

    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    headlines = []
    time = []
    newslink = []

    for i in soup.find_all('a', class_ = 'LatestNews-headline'):
        headlines.append(i.text)

    for i in soup.find_all('time', class_ = 'LatestNews-timestamp'):
        time.append(i.text)

    for i in soup.find_all('a', {'class' : 'LatestNews-headline'}):
        newslink.append(i.get('href'))
    ind = []
    for i in range(1,31):
        ind.append(i)
    
    cnbc_headlines = pd.DataFrame({'Headlines': headlines, 'Published time': time, 'Headlines_link': newslink}, index = ind)
    
    return cnbc_headlines

```

```python
url = 'https://www.cnbc.com/world/?region=world'
headline = latest(url)
headline
```
