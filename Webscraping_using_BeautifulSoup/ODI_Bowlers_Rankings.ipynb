```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def ODI_bowlers_ranking(url):
    
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    a = soup.find('div', class_ = 'rankings-block__banner--name-large').text
    players = []
    for i in soup.find_all('td', class_ = 'table-body__cell rankings-table__name name', limit = 9):
        players.append(i.text.strip('\n'))
    players.insert(0,a)

    b = soup.find('div', class_ = 'rankings-block__banner--nationality').text[2:5]
    team = []
    for i in soup.find_all('span', class_ = 'table-body__logo-text', limit = 9):
        team.append(i.text)
    team.insert(0,b)

    c = soup.find('div', class_ = 'rankings-block__banner--rating').text
    rating = []
    for i in soup.find_all('td', class_ = 'table-body__cell rating', limit = 9):
        rating.append(i.text)
    rating.insert(0,c)
    
    ind = []
    for i in range(1,11):
        ind.append(i)

    odi_bowlers_ranking = pd.DataFrame({'Player': players, 'Team': team, 'Rating': rating}, index = ind)
    return odi_bowlers_ranking
```

```python
url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling'
ODI_bowlers_ranking = ODI_bowlers_ranking(url)
ODI_bowlers_ranking
```
