```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup 
```

```python
def restuarants(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser' )

    hotel = []
    for i in soup.find_all('a', class_ = 'restnt-name ellipsis', limit = 19):
        hotel.append(i.text)
    cuisine = []
    for i in soup.find_all('span', class_ = 'double-line-ellipsis', limit = 19):
        cuisine.append(i.find('a').text)
    address = []
    for i in soup.find_all('div', class_ = 'restnt-loc ellipsis', limit = 19):
        address.append(i.text)
    ratings = []
    for i in soup.find_all('div', class_ = 'restnt-rating rating-4'):
        ratings.append(i.text)
    images = []
    for i in soup.find_all('img', class_="no-img", limit = 19):
        images.append(i['data-src'])
    
    ind = []
    for i in range(1,20):
         ind.append(i)

    Dine_in = pd.DataFrame({'Restaurant name': hotel, 'Cuisine': cuisine, 'Location': address, 'Ratings': ratings, 'Image URL':images}, index = ind)

    return Dine_in
```

```python
url = 'https://www.dineout.co.in/bangalore-restaurants'
res = restuarants(url)
res
```
