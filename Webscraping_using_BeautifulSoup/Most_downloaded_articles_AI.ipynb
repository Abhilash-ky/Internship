```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def journal(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    article = []
    for i in soup.find_all('h2', class_ = 'sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'):
        article.append(i.text)
    authors = []
    for i in soup.find_all('span', class_ = 'sc-1w3fpd7-0 dnCnAO'):
        authors.append(i.text)
    year = []
    for i in soup.find_all('span', class_ = 'sc-1thf9ly-2 dvggWt'):
        year.append(i.text)
    ind = []
    for i in range(1,26):
        ind.append(i)
    most_downloaded_articles = pd.DataFrame({'Article': article, 'Authors': authors, 'Year of Publishing': year}, index = ind)
    
    return most_downloaded_articles
    
```

```python
url = 'https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles'
df = journal(url)
df
```
