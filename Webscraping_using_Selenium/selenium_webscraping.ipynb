### Q1. Python program to scrape data for “Data Analyst” Job position in “Bangalore” location.

```python
driver = drive()
url = 'https://www.naukri.com/'
driver.get(url)
driver.maximize_window()
desig = driver.find_element(By.CLASS_NAME, 'suggestor-input ')
desig.send_keys('Data Analyst')
location = driver.find_element(By.XPATH, "/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys('Bangalore')
search = driver.find_element(By.CLASS_NAME, "qsbSubmit")
search.click()
time.sleep(5)
```

```python
job_title = []
job_location = []
company_name = []
experience_required = []
title = driver.find_elements(By.XPATH, '//*[@class ="title ellipsis"]')
for i in title[0:10]:
    t = i.text
    job_title.append(t)
location = driver.find_elements(By.XPATH,'//*[@class = "ellipsis fleft locWdth"]')
for i in location[0:10]:
    l = i.text
    job_location.append(l)
company = driver.find_elements(By.XPATH, '//*[@class = "subTitle ellipsis fleft"]')
for i in company[0:10]:
    c = i.text
    company_name.append(c)
experience = driver.find_elements(By.XPATH, '//*[@class = "ellipsis fleft expwdth"]')
for i in experience[0:10]:
    e = i.text
    experience_required.append(e)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
import pandas as pd
index1 = []
for i in range(1,11):
    index1.append(i)
data_analyst_roles= pd.DataFrame({'Title':job_title, 'Location': job_location, 'Company':company_name, 'Experience':experience_required}, index = index1)
data_analyst_roles
```

### Q4. Python program to scrape data of sunglasses from flipkart.com and store into dataframe

```python
driver = drive()
url = 'https://www.flipkart.com/'
driver.get(url)
driver.find_element(By.XPATH,'/html/body/div[2]/div/div/button').click()
driver.find_element(By.XPATH, '//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input').send_keys('sunglasses')
driver.find_element(By.XPATH, '//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button').click()
driver.implicitly_wait(10)
Brand = []
Product_Description = []
Price = []
discount = []

for i in range(3):
    
    
    brand = driver.find_elements(By.XPATH,"//div[@class='_2WkVRV']")
    product_des = driver.find_elements(By.XPATH,"//div[@class='_2B099V']/a[1]")
    price = driver.find_elements(By.XPATH,"//div[@class='_30jeq3']")
    disc = driver.find_elements(By.XPATH, "//div[@class='_3Ay6Sb']")


    for i in brand:
        Brand.append(i.text)
    for i in product_des:
        Product_Description.append(i.text)
    for i in price :
        Price.append(i.text)
    for i in disc:
        discount.append(i.text)
    
driver.implicitly_wait(10)

nxt_button = driver.find_element(By.XPATH,"//a[@class='_1LKTO3']")
nxt_button.click()
```

```python
Brands = Brand[:100]
Prod = Product_Description[:100]
price = Price[:100]
discounted = discount[:100]
```

```python
indexing = []
for i in range(1,101):
    indexing.append(i)
sunglasses = pd.DataFrame({'Brand':Brands, 'Product Description': Prod, 'Price': price, 'Discount offered': discounted}, index = indexing)
sunglasses
```

### Q5. Python program to scrape data of reviews from flipkart.com for iphone 11 64 gb

```python
driver = drive()
url = 'https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART&page='
driver.get(url)
driver.implicitly_wait(10)
ratings = []
rev = []
full = []
for page in range(1,11):
    
    rate = driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div/div[1]/div[2]/div/div/div/div/div[1]/div')
    review = driver.find_elements(By.XPATH, '//*[@id="container"]/div/div[3]/div/div[1]/div[2]/div/div/div/div/div[1]/p')
    summ = driver.find_elements(By.XPATH, '//*[@id="container"]/div/div[3]/div/div[1]/div[2]/div/div/div/div/div[2]/div/div/div')
    driver.implicitly_wait(10)
    for i in rate[2:]:
        ratings.append(i.text)
    
    
    for i in review:
        rev.append(i.text)
    
    
    
    for i in summ:
        full.append(i.text)


driver.implicitly_wait(10)
nxt_button = driver.find_element(By.XPATH,"//a[@class='ge-49M']")
nxt_button.click()

    
```

```python
indexing = []
for i in range(1,101):
    indexing.append(i)
iphone_listing = pd.DataFrame({'Ratings': ratings, 'Review Summary' : rev, 'Full review' : full}, index = indexing)
iphone_listing
```

### Q8. Pyhton program scrape data from azqoutes.com and create dataframe of top 100 qoutes

### Q9. Python program to scrape data of list of all prime ministers using selenium from jagranjosh.com

```python
driver = drive()
url = 'https://www.jagranjosh.com/'
driver.get(url)
driver.find_element(By.XPATH, '/html/body/div/div[1]/div/div[1]/div/div[5]/div/div[1]/header/div[3]/ul/li[9]/a').click()
a = driver.find_element(By.XPATH, '/html/body/div/div/div/div[2]/div/div[5]/div/div/ul/li[2]/a/i')
a.click()
driver.find_element(By.XPATH, '/html/body/div/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a').click()
prime_ministers = []
prime = driver.find_elements(By.XPATH, '//*[@id="itemdiv"]/div[4]/span/div[2]/table/tbody/tr/td[2]/p')
for i in prime:
    p = i.text
    prime_ministers.append(p)
period = []
time = driver.find_elements(By.XPATH, '//*[@id="itemdiv"]/div[4]/span/div[2]/table/tbody/tr/td[3]/p')
for i in time:
    t = i.text
    period.append(t)
timeline = [] 
term = driver.find_elements(By.XPATH, '//*[@id="itemdiv"]/div[4]/span/div[2]/table/tbody/tr/td[4]')
for i in term:
    tp = i.text
    timeline.append(tp)
remarks = []
remark = driver.find_elements(By.XPATH, '//*[@id="itemdiv"]/div[4]/span/div[2]/table/tbody/tr/td[5]/p')
for i in remark:
    r = i.text
    remarks.append(r)
```

```python
print(len(prime_ministers),len(period), len(timeline), len(remarks))
```

```python
indexing = []
for i in range(1,19):
    indexing.append(i)
primeministers_of_India = pd.DataFrame({'Name': prime_ministers, 'Born-dead': period, 'Term of office': timeline, 'Remarks' : remarks}, index = indexing)

primeministers_of_India
```
